```python
import os
import json
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import figure

data = []
names = []
for file in os.listdir("LOGTIME"):
    csv = pd.read_csv("./LOGTIME/"+file)
    c = np.array(csv)
    data.append(c[:,0])
    name = os.path.basename("./LOGTIME/"+file)
    name = name[5:]
    name = name[:-4]
    names.append(name)
    
figure(num=None, figsize=(12, 7), dpi=80, facecolor='w', edgecolor='k')

mean = np.array([np.mean(e) for e in data])
std = np.array([np.std(e, ddof=1) for e in data])

new_mean = pd.DataFrame(names)
new_mean["mean"] = mean
new_mean.sort_values(by=[0])
mean = np.array(new_mean.sort_values(by=[0]))
mean

new_mean = pd.DataFrame(names)
new_mean["std"] = std
new_mean.sort_values(by=[0])
std = np.array(new_mean.sort_values(by=[0]))
std

mean = np.delete(mean, 10,0)
mean = np.delete(mean, 10,0)
mean = np.delete(mean, 10,0)
mean = np.delete(mean, 15,0)
mean = np.delete(mean, 15,0)

std = np.delete(std, 10,0)
std = np.delete(std, 10,0)
std = np.delete(std, 10,0)
std = np.delete(std, 15,0)
std = np.delete(std, 15,0)

plt.plot(mean[:,0], mean[:,1], color='yellow', marker='o',linewidth=3, markersize=10, label="Média")

plt.ylabel('Tempo de execução (microssegundos)',fontsize=14)
plt.xlabel('Testes',fontsize=14)

plt.legend()
plt.show()
```

```python
figure(num=None, figsize=(12, 7), dpi=80, facecolor='w', edgecolor='k')

plt.plot(std[:,0], std[:,1], color='red', marker='o',linewidth=3, markersize=10, label="Desvio padrão")

plt.ylabel('Tempo de execução (microssegundos)',fontsize=14)
plt.xlabel('Número de ilhas',fontsize=14)

plt.legend()
plt.show()
```
